Simple function to extract seeds from a grayscale image painted with Paint (red for object seeds/ blue for background seeds).

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
def ST_from_init(img_ST):
    red, green, blue = np.transpose(img_ST, (2,0,1))
    S_mask = (red > 200) * (green < 100) * (blue < 100)
    T_mask = (red < 80) * (green < 80) * (blue > 140)
    S_indices = np.argwhere(S_mask)
    T_indices = np.argwhere(T_mask)
    return S_indices, T_indices, S_mask, T_mask
```

```python
img = np.array(Image.open("rose.png").convert("L"))
img_ST = np.array(Image.open("rose_ST.png").convert("RGB"))

S_indices, T_indices, S_mask, T_mask = ST_from_init(img_ST)
img_retrieve = np.copy(img_ST)
img_retrieve[S_mask] = (100,0,100)
img_retrieve[T_mask] = (0,100,0)

plt.figure(figsize=(20,10))

plt.subplot(131)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(132)
plt.title("Painted S & T")
plt.imshow(img_ST)
plt.axis('off')

plt.subplot(133)
plt.title("Retrieved S & T")
plt.imshow(img_retrieve)
plt.axis('off')

plt.show()
```
